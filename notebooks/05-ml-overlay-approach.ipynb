```python
import pandas as pd
import os
```

```python
# Define file paths
office_features_path = "../features/office/extracted-features.csv"
nature_features_path = "../features/nature/extracted-features.csv"
```

```python
# Load features
office_df = pd.read_csv(office_features_path)
nature_df = pd.read_csv(nature_features_path)

# Inspect
print("Office features:", office_df.shape)
print("Nature features:", nature_df.shape)
print(office_df.head())
```

# Step 2: Filter Ambient Nature Segments

```python
# Basic ambient filtering criteria
ambient_df = nature_df[
    (nature_df['rms'] < 0.02) &
    (nature_df['onset_density'] < 1.5) &
    (nature_df['zcr'] < 0.1)
].copy()

# Optional: sanity check
print(f"Original nature segments: {len(nature_df)}")
print(f"Ambient nature segments: {len(ambient_df)}")
ambient_df.head()
```

# Step 3: Cluster Office Audio

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load office features
office_df = pd.read_csv("features/office/office_features.csv")

# Select relevant columns
feature_cols = ['rms', 'centroid', 'flatness', 'bandwidth', 'onset_density']
X = office_df[feature_cols].values

# Standardize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply KMeans
k = 5  # You can experiment with different k values
kmeans = KMeans(n_clusters=k, random_state=42)
labels = kmeans.fit_predict(X_scaled)

# Save back to dataframe
office_df['cluster'] = labels

# Optional: see cluster sizes
print(office_df['cluster'].value_counts())
office_df.head()
```
